```python
import pandas as pd
df = pd.read_csv('./music.csv')
```

```python
df.head()
df.shape[0]
```

```python
df['date'] = pd.to_datetime(df['date'], format='mixed', dayfirst=True)

# Extract year from the date and create a new column
df['year'] = df['date'].dt.year

# Group by year and find the min and max date for each year
date_ranges_per_year = df.groupby('year')['date'].agg(['min', 'max'])

print(date_ranges_per_year)
```

```python
df['date'] = pd.to_datetime(df['date'])

# Extract year from 'date' if not already done
df['year'] = df['date'].dt.year

# Group by 'year' and list all dates
for year, group in df.groupby('year'):
    print(f"Year {year}:")
    # Get all unique dates within this year, sorted
    unique_dates = group['date'].sort_values().unique()
    # Print or process the unique dates
    for date in unique_dates:
        print(date)

    print("\n")  # Newline for readability between years
```

```python
print(df['position'].describe())
unique_positions = df['position'].unique()
print(sorted(unique_positions))
```

```python
import matplotlib.pyplot as plt
df['position'].hist(bins=50)
plt.title('Distribution of Positions')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.show()
```
